```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import scale
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import fcluster
```

```python
# Update the path for importing the data
df = pd.read_csv("C:\\Users\\MOUMITA\\Desktop\\Book2.csv")
df.head()
```

```python
# Select three variables for clustering purpose
df1 = df.iloc[:, 1:3]
df1.head()
```

```python
# Standardization of features (Feature scaling)
scaler = StandardScaler()
X = scaler.fit_transform(df1)
X
```

```python
# Fitting two clusters
kmeans = KMeans(n_clusters=4, n_init=10, random_state=0)  # create a k-means object
kmeans.fit(X)
```

```python
# Cluster membership
labels = kmeans.predict(X)
labels
```

```python
# Cluster description
df1['cluster'] = labels  # add cluster membership number as new column
df1['company'] = df['Company Name']
df1
```

```python
# Clusters
for i in range(4):
    print(f"Cluster {i}:")
    print(df1[df1['cluster'] == i])

df = df1[df1['cluster'] == 1]
df.drop('cluster', axis=1, inplace=True)

import random
from deap import base, creator, tools, algorithms

companies = {
    'Axis Bank': (0.000199, 0.025445),
    'Eicher Motors': (0.000387, 0.021775),
    'HDFC Life': (0.000315, 0.020703),
    'M&M': (0.000572, 0.022309),
    'Maruti Suzuki': (0.000177, 0.021146),
    'ONGC': (0.000013, 0.025678),
    'Tech Mahindra': (0.000278, 0.020248),
    'UPL': (0.000211, 0.024154)
}

POPULATION_SIZE = 50
NUM_GENERATIONS = 100
CXPB = 0.5
MUTPB = 0.2

def evaluate_portfolio(individual):
    total_return = sum(companies[company][0] for company, included in zip(companies.keys(), individual) if included)
    total_risk = sum(companies[company][1] for company, included in zip(companies.keys(), individual) if included)
    return total_return, total_risk

creator.create("FitnessMax", base.Fitness, weights=(1.0, -1.0))
creator.create("Individual", list, fitness=creator.FitnessMax)

toolbox = base.Toolbox()
toolbox.register("attr_bool", random.randint, 0, 1)
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_bool, len(companies))
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

toolbox.register("evaluate", evaluate_portfolio)
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutFlipBit, indpb=0.05)
toolbox.register("select", tools.selNSGA2)

population = toolbox.population(n=POPULATION_SIZE)

for gen in range(NUM_GENERATIONS):
    offspring = algorithms.varAnd(population, toolbox, cxpb=CXPB, mutpb=MUTPB)
    fits = toolbox.map(toolbox.evaluate, offspring)
    for fit, ind in zip(fits, offspring):
        ind.fitness.values = fit

    population = toolbox.select(offspring, k=len(population))

best_portfolio = tools.selBest(population, k=1)[0]

best_portfolio_companies = [company for company, included in zip(companies.keys(), best_portfolio) if included]
best_portfolio_return, best_portfolio_risk = evaluate_portfolio(best_portfolio)

print("Best Portfolio:", best_portfolio_companies)
print("Average Rate of Return:", best_portfolio_return)
print("Risk (Standard dev.):", best_portfolio_risk)
```

```python
# Mean-Variance Optimization for Portfolio Weights
import cvxopt as opt
from cvxopt import blas, solvers

returns = np.array([companies[company][0] for company in best_portfolio_companies])
risks = np.array([companies[company][1] for company in best_portfolio_companies])

n = len(best_portfolio_companies)

# Convert to cvxopt matrices
S = opt.matrix(np.diag(risks))
pbar = opt.matrix(returns)

# Create constraint matrices
G = -opt.matrix(np.eye(n))   # negative n x n identity matrix
h = opt.matrix(0.0, (n ,1))
A = opt.matrix(1.0, (1, n))
b = opt.matrix(1.0)
```

```python
# Calculate optimal weights
sol = solvers.qp(S, -pbar, G, h, A, b)
weights = np.array(sol['x']).flatten()

print("Optimal Weights for Minimizing Risk:")
for company, weight in zip(best_portfolio_companies, weights):
    print(f"{company}: {weight:.4f}")
```
